```python
import pandas as pd

DADOS = '../dados/ml_project1_data.csv'
                                                                                                                                                                                                                                                                    
df = pd.read_csv(DADOS, sep='\t')

df.head()
```

```python
df.info()
```

```python
# Retirando os valores nulos
df = df.dropna()

# Convertendo a coluna Dt_Customer para datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')

# motrar as primeiras linhas
df.head()
```

```python
df.info()
```

```python
df['Dt_Customer'].describe()
```

```python
df['DaysSinceEnrolled'] = df['Dt_Customer'].max() - df['Dt_Customer']
df['DaysSinceEnrolled'] = df['DaysSinceEnrolled'].dt.days

df.head()
```

```python
df['YearsSinceEnrolled'] = df['DaysSinceEnrolled'] // 365
df.head()
```

```python
df['YearsSinceEnrolled'].value_counts()
```

```python
df['Age'] = df['Dt_Customer'].max().year - df['Year_Birth']

df.head()
```

```python
df['Age'].describe()
```

```python
# grupos de idade
# adulto_jovem 18-30
# adulto 31-45
# adulto_senior 46-60
# idoso 61+

df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[18, 30, 45, 60, df['Age'].max()],
    include_lowest=True,
    labels=['adulto_jovem', 'adulto', 'adulto_senior', 'idoso']
)
```
